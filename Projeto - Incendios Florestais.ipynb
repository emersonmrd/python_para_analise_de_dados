```python
# Libs Necess치rias 

# Libs para modelagem e matrizes
import numpy as np
import pandas as pd

# Libs para an치lises gr치ficas
import matplotlib.pyplot as plt
import seaborn as sns

# Libs para ignorar avisos
import warnings

# Desabilitando os avisos
warnings.filterwarnings('ignore')
```

```python
# Lendo os dados
Base_Dados = pd.read_csv('datasets/Dados_Indendio.csv', encoding='latin-1')

# Verificando
Base_Dados.head()
```

```python
# Selecionar apenas as colunas num칠ricas, excluindo a coluna 'year'
colunas_numericas = Base_Dados.select_dtypes(include=['number']).columns.difference(['year'])

# Analise de numero de incendios por ano
Analise = Base_Dados.groupby(by=['year'])[colunas_numericas].sum().reset_index()

# Tamanho
plt.figure( figsize=(12,5) )

# Style
plt.style.use('ggplot')

# Grafico
plt.title( 'Total de inc칡ndios no Brasil: 1997 - 2017', loc='left', fontsize=14 )
sns.lineplot( data=Analise, x='year', y='number', estimator='sum', lw=2, color='#ff5555', alpha=0.85)

# Labels
plt.xlabel('Quantidade')
plt.ylabel('Per칤odo');
```

```python
# analise por mes do numero de incendios
Analise_02 = Base_Dados.groupby(by=['year', 'month'])[colunas_numericas].sum().reset_index()
Analise_02.head()

# Tamanho
plt.figure( figsize=(12,5) )

# Gr치fico
plt.title( 'Inc칡ndios por m칡s', loc='left', fontsize=14 )
sns.boxplot( data=Analise_02, x='month', y='number', palette='coolwarm', saturation=1, width=0.9, linewidth=2,
            order=['Janeiro', 'Fevereiro', 'Mar칞o', 'Abril', 'Maio', 'Junho', 'Julho', 'Agosto', 'Setembro', 'Outubro', 'Novembro', 'Dezembro'] )

# Labels
plt.xlabel('M칡s')
plt.ylabel('N칰mero de inc칡ndios');

```

```python
# analise de numero de incendios por estado
Analise_03 = Base_Dados.groupby(by=['state'])[colunas_numericas].sum().reset_index().sort_values( 'number',ascending=False )
Analise_03.head()

# Tamanho
plt.figure( figsize=(12,5) )

# Gr치fico
plt.title( 'Estados com maior n칰mero de inc칡ndios', loc='left', fontsize=14 )
plt.bar( Analise_03.state, Analise_03['number'], color='#f44e3f' )

# Labels
plt.ylabel('Quantidade')
plt.xticks( rotation=90);
```

```python
# Estados TOP 10
Lista_TOP10 = Analise_03['state'][0:10].values

# Tamanho
plt.figure( figsize=(12, 5) )

# Grafico
plt.title( 'Top 10 Estados com inc칡ndios', loc='left', fontsize=14 )

# Loop
for Coluna in Lista_TOP10:
  
  # Filtrar o estado
  Filtro = Base_Dados.loc[ Base_Dados['state'] == Coluna ]

  # Agrupar os valores para sumarizar
  Analise_Local = Filtro.groupby( by=['year'] ).sum().reset_index()

  # Plot
  sns.lineplot(data=Analise_Local, x='year', y='number', lw=2, alpha=0.85)

# Labels
plt.xlabel('Per칤odo')
plt.ylabel('N칰mero de inc칡ndios')

# Legenda
plt.legend( Lista_TOP10, bbox_to_anchor=(1, 0.7) );
```
