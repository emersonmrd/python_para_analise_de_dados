```python
# Libs Necessárias

# Libs para Modelagem e Matrizez
import numpy as np
import pandas as pd

# Libs para análises gráficas
import matplotlib.pyplot as plt
import seaborn as sns

# Lib para ignorar avisos
import warnings

# Desabilitando avisos
warnings.filterwarnings('ignore')
```

```python
# Lendo os dados
Base_Dados = pd.read_csv('datasets/PS4_GamesSales.csv', encoding='latin-1')

# Verificando
Base_Dados.head()
```

```python
# Dimensão
Base_Dados.shape
```

```python
# Nulos
Base_Dados.isnull().sum()
```

```python
# Nulos Graficos
plt.figure( figsize=(14,5) )
plt.title('Verificando campos nulos')
sns.heatmap( Base_Dados.isnull(), cbar=False );
```

```python
# Retirando os nulos
Base_Dados.dropna( inplace=True )
```

```python
# Estatisticas 
Base_Dados.describe()
```

```python
# Tamanho da imagem
plt.figure( figsize=(10,5) )

# Titulo
plt.title('Quantidade de Vendas Globais (mi)', loc='left', fontsize=14)

# Grafico
sns.barplot( data=Base_Dados, x='Year', y='Global', ci=None, color='#69b3a2', estimator=sum)

# Label
plt.ylabel('Quantidade de Vendas Globais (mi)');
```

```python
# Retirar os anos
Base_Dados = Base_Dados.loc[ (Base_Dados['Year'] != 2019) & (Base_Dados['Year'] != 2020)]

Base_Dados.head()
```

```python
# Tamanho

plt.figure( figsize=(15, 5) )

# Estilo
plt.style.use('ggplot')

# titulo
plt.title('Distribuição das Vendas Globais', loc='left', fontsize=14 )

# Plot
sns.kdeplot( Base_Dados['Global'], shade=True, bw=1, color='#96a8a8', linewidth=2.5);
```

```python
Base_Dados.groupby( by=['Year'] ).sum()
```

```python
# Tamanho
plt.figure( figsize=(10, 4) )

# titulo
plt.title('Análise da distribuição Global (mi)')

# Label
plt.ylabel('Quantidade de Vendas Globais (mi)');

# Plot
sns.boxplot( data=Base_Dados, x='Year', y='Global');
```

```python
Base_Dados.loc[ Base_Dados['Global'] >= 10]
```

```python
Base_Dados
```

```python
Analise = Base_Dados.groupby( by=['Year'] ).sum().reset_index()

# Analisando a proporção dos 100% de cada continente comparado ao total de vendas
America = [ America / Total * 100 for America, Total in zip( Analise['North America'] , Analise['Global'] ) ]
Europa = [ Europa / Total * 100 for Europa, Total in zip( Analise['Europe'] , Analise['Global'] ) ]
Japao = [ Japao / Total * 100 for Japao, Total in zip( Analise['Japan'] , Analise['Global'] ) ]
Mundo = [ Mundo / Total * 100 for Mundo, Total in zip( Analise['Rest of World'] , Analise['Global'] ) ]

America, Europa, Japao, Mundo
```

```python
Analise['Year'].unique()
```

```python
# Tamanho
plt.figure( figsize=(10, 5) )

# Largura barra no gráfico
Largura_Barra = 0.85

Rotulos = Analise['Year']
Grupos = [0, 1, 2, 3, 4, 5]

# titulo
plt.title('Análise da distribuição por continentes')

# Plot da America
plt.bar( Grupos, America, width=Largura_Barra, color='#b5ffb9', edgecolor='white')

# Plot da Europa
plt.bar( Grupos, Europa, bottom=America, width=Largura_Barra, color='#f9bc86', edgecolor='white')

# Plot do Japao
plt.bar( Grupos, Japao, bottom=[ A + B for A, B in zip( America, Europa ) ], width=Largura_Barra, color='#a3acff', edgecolor='white')

# Plot do Resto do mundo
plt.bar( Grupos, Mundo, bottom=[ A + B + C for A, B, C in zip( America, Europa, Japao ) ], width=Largura_Barra, color='#d3acfe', edgecolor='white')

# Labels
plt.xticks( Grupos, Rotulos )
plt.xlabel('Grupo')
plt.ylabel('Destribuição %')

# Legenda 
plt.legend(['America N', 'Europa', 'Japão', 'Mundo'], loc='upper left', bbox_to_anchor=(0.15, -0.1), ncol=4 );
```

```python
from sklearn.preprocessing import LabelEncoder
Funcao_Label = LabelEncoder()

Base_Dados['Produtor'] = Funcao_Label.fit_transform( Base_Dados['Publisher'] )
Base_Dados['Genero'] = Funcao_Label.fit_transform( Base_Dados['Genre'] )
Base_Dados['Jogo'] = Funcao_Label.fit_transform( Base_Dados['Game'] )

Base_Dados.head()
```

```python
Paleta_Cores = sns.color_palette('husl', 8)
Paleta_Cores
```

```python
plt.figure( figsize=(20, 5))
plt.title('Análise dos produtores de Game (mi)')
sns.scatterplot(data=Base_Dados, x='Produtor', y='Global', color=Paleta_Cores[0]);
```

```python
plt.figure( figsize=(20, 5))
plt.title('Análise dos generos dos Games (mi)')
sns.scatterplot(data=Base_Dados, x='Genero', y='Global', color=Paleta_Cores[0]);
```

```python
plt.figure( figsize=(20, 5))
plt.title('Análise dos Games (mi)')
sns.scatterplot(data=Base_Dados, x='Jogo', y='Global', color=Paleta_Cores[0]);
```

```python
fig, axs = plt.subplots(figsize=(18, 15), nrows=3, ncols=2)

# Cor de fundo
Cor_Fundo = '#f5f5f5'
for ax in axs.flat:
    ax.set_facecolor(Cor_Fundo)
fig.set_facecolor(Cor_Fundo)

# Estilo dos gráficos
plt.style.use('seaborn-v0_8-darkgrid')

# Titulo da figura
plt.suptitle('Python para Análise de Dados \n Projeto prático 5 - Análise Mercado de Games PS4', fontsize=22, color='#404040', fontweight=600)

# Acessando o gráfico 1
axs[0, 0].set_title('Quantidade de Vendas Globais (mi)', loc='left', fontsize=14)
axs[0, 0].bar(Base_Dados['Year'], Base_Dados['Global'], color='#69b3a2')
axs[0, 0].set_ylabel('Quantidade de Vendas Globais (mi)')
axs[0, 0].tick_params(axis='x', rotation=45)  

# Acessando o gráfico 2
axs[0, 1].set_title('Análise da distribuição Global (mi)', loc='left', fontsize=14)
axs[0, 1].set_ylabel('Quantidade de Vendas Globais (mi)')
sns.boxplot(data=Base_Dados, x='Year', y='Global', ax=axs[0, 1])
axs[0, 1].tick_params(axis='x', rotation=45)  

# Acessando o gráfico 3
axs[1, 0].set_title('Análise da distribuição por continentes', loc='left', fontsize=14)
axs[1, 0].bar(Grupos, America, width=0.85, color='#b5ffb9', edgecolor='white')
axs[1, 0].bar(Grupos, Europa, bottom=America, width=0.85, color='#f9bc86', edgecolor='white')
axs[1, 0].bar(Grupos, Japao, bottom=[A + B for A, B in zip(America, Europa)], width=0.85, color='#a3acff', edgecolor='white')
axs[1, 0].bar(Grupos, Mundo, bottom=[A + B + C for A, B, C in zip(America, Europa, Japao)], width=0.85, color='#d3acfe', edgecolor='white')
axs[1, 0].set_xticks(Grupos)
axs[1, 0].set_xticklabels(Rotulos, rotation=45)
axs[1, 0].set_xlabel('Grupo')
axs[1, 0].set_ylabel('Distribuição %')
axs[1, 0].legend(['America N', 'Europa', 'Japão', 'Mundo'], loc='upper left', bbox_to_anchor=(0.15, -0.2), ncol=4)

# Acessando o gráfico 4
axs[1, 1].set_title('Análise dos produtores de Game (mi)', loc='left', fontsize=14)
sns.scatterplot(data=Base_Dados, x='Produtor', y='Global', color=Paleta_Cores[0], ax=axs[1, 1])
axs[1, 1].tick_params(axis='x', rotation=45)  

# Acessando o gráfico 5
axs[2, 0].set_title('Análise dos gêneros dos Games (mi)', loc='left', fontsize=14)
sns.scatterplot(data=Base_Dados, x='Genero', y='Global', color=Paleta_Cores[0], ax=axs[2, 0])
axs[2, 0].tick_params(axis='x', rotation=45) 

# Acessando o gráfico 6
axs[2, 1].set_title('Análise dos Games (mi)', loc='left', fontsize=14)
sns.scatterplot(data=Base_Dados, x='Jogo', y='Global', color=Paleta_Cores[0], ax=axs[2, 1])
axs[2, 1].tick_params(axis='x', rotation=45)  

# Ajustar o layout
plt.subplots_adjust(hspace=0.45, wspace=0.15)

# Rodapé
Rodape = '''
Esse relatório foi elaborado no treinamento "Python para Análise de Dados"
Está disponível no canal do youtube @Data Viking
by: @Emerson Martins
'''

# Incluindo o rodapé no relatório
fig.text(0.5, -0.02, Rodape, ha='center', va='bottom', size=12, color='#938ca1')

plt.show()
```
