```python
# Bibliotecas
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Lendo a base de dados
Base_Dados = pd.read_excel('datasets/Dados_Pib.xlsx')

# Verificando 
Base_Dados.head(10)
```

```python
Base_Dados.groupby( by=['Territorialidades', 'Ano'] ).mean()
```

```python
#  Sistemas de Grids

# Cor de Fundo
Cor_fundo = '#f5f5f5'

# Incluir cor fundo


# Criar p sistema de Grids
Grid_Graficos = sns.FacetGrid( Base_Dados, col='Territorialidades', hue='Territorialidades', col_wrap=4)

# Adicionar o grafico de linhas em cada gráfico 
Grid_Graficos = Grid_Graficos.map( plt.plot, 'Ano', 'PIB per capita')

# Adicionar uma sombra + Ajuste do titulo
Grid_Graficos = Grid_Graficos.map( plt.fill_between, 'Ano', 'PIB per capita', alpha=0.2 ).set_titles('{col_name} Territorialidades')

# Filtrar o titulo
Grid_Graficos = Grid_Graficos.set_titles('{col_name}')

# Adicionar um subtitulo
Grid_Graficos = Grid_Graficos.figure.suptitle(
    'Evolução da Renda per capita por Estado \n Essse relatório foi elaborado no treinamento "Python para Análise de Dados" \n Está dispónivel no canal do youtube @Data Viking',
    fontsize=18
)

# Ajustando 
plt.subplots_adjust( top=0.92 )
```
