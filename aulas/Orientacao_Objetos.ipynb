# **Programação orientada a Objetos in Python**

A programação orientada a objetos (POO) é um paradigma de programação que organiza o código em torno de "objetos", que podem conter dados na forma de campos, também conhecidos como atributos, e código na forma de procedimentos, também conhecidos como métodos. Esses objetos são então utilizados para representar entidades do mundo real ou conceitos abstratos no software.

## **Existem quatro princípios fundamentais na POO**

image.png

**Abstração**

Abstração é o processo de focar nos aspectos essenciais de um objeto enquanto oculta os detalhes menos relevantes. Isso permite que os desenvolvedores vejam apenas o que é relevante para a sua interação com os objetos, sem precisar se preocupar com como eles são implementados internamente.



**Encapsulamento**

Encapsulamento é o conceito de agrupar os dados e os métodos que operam nesses dados em uma única unidade, chamada de classe. Isso impede o acesso direto aos dados, exigindo que o código externo utilize métodos específicos para interagir com eles. Isso ajuda a proteger os dados de modificações não autorizadas e facilita a manutenção do código.



**Herança**

Herança é o mecanismo pelo qual uma classe pode herdar os campos e métodos de outra classe, conhecida como classe pai ou superclasse. Isso promove a reutilização de código e permite a criação de hierarquias de classes, onde as classes mais específicas (subclasses) herdam comportamentos e características das classes mais gerais (superclasses).



**Polimorfismo**

Polimorfismo é a capacidade de objetos de classes diferentes responderem de maneiras diferentes a mensagens ou chamadas de método com o mesmo nome. Isso permite que um único método tenha diferentes implementações em diferentes classes, facilitando a programação genérica e a extensibilidade do código.



## **Programação orientada a objetos**

Python é uma linguagem de programação que suporta totalmente POO, permitindo a criação de classes, objetos, herança, encapsulamento e polimorfismo. Aqui estão alguns conceitos chave sobre o uso de POO em Python.

**Classes e Objetos**: Em Python, uma classe é uma estrutura que define um tipo de objeto. Ela é como um plano para criar objetos. Um objeto é uma instância de uma classe. Por exemplo, podemos ter uma classe Cachorro que define características e comportamentos de um cachorro, e então podemos criar objetos individuais dessa classe, como meu_cachorro = Cachorro().

**Atributos e Métodos**: As classes em Python podem conter atributos (variáveis) e métodos (funções). Atributos representam os dados associados a um objeto, enquanto métodos representam as ações que um objeto pode realizar. Por exemplo, em uma classe Cachorro, podemos ter atributos como nome, idade e métodos como latir() e andar().

**Herança**: Python suporta herança, permitindo que uma classe herde atributos e métodos de outra classe. Isso promove a reutilização de código e a criação de hierarquias de classes. Por exemplo, podemos ter uma classe Animal com atributos e métodos genéricos, e então criar classes mais específicas, como Cachorro e Gato, que herdam da classe Animal.

**Encapsulamento**: Embora Python não tenha modificadores de acesso como em outras linguagens de programação, é possível alcançar encapsulamento por convenção. Atributos e métodos que começam com um sublinhado (_) são considerados "privados" e devem ser acessados apenas dentro da própria classe. Isso ajuda a proteger os dados e promove uma melhor organização do código.

**Polimorfismo**: Python suporta polimorfismo, permitindo que métodos com o mesmo nome se comportem de maneiras diferentes em classes diferentes. Isso permite escrever código genérico que pode lidar com diferentes tipos de objetos sem a necessidade de saber antecipadamente o tipo específico de cada objeto.

````python
# Criar nossa Classe
class Alguem:

    ### METODO CONSTRUTOR ###
    def __init__(self, nome, idade):
        self.nome = nome
        self.idade = idade

    def imprimir_informacoes(self):
        ''' Função para mostrar o nome e idade da pessoa'''
        print(f"Nome: {self.nome}, Idade: {self.idade}")
````


O método __init__ é um método especial em Python usado para inicializar objetos de uma classe. Ele é chamado de "construtor" porque é executado automaticamente sempre que uma nova instância da classe é criada.
<br>
<br>
def __init__(self, parametro1, parametro2, ...): Define o método __init__. Ele sempre recebe self como seu primeiro parâmetro, que é uma referência à instância atual da classe. Além disso, ele pode receber outros parâmetros que são passados quando uma instância da classe é criada.
<br>
<br>
**self**: Representa a instância atual da classe. É usado dentro do método para acessar e modificar os atributos e métodos da instância.
<br>
<br>
**parametro1, parametro2**, ...: São os parâmetros que você deseja passar para inicializar os atributos do objeto. Você pode passar quantos parâmetros forem necessários para inicializar os atributos da classe.
<br>
<br>
Dentro do método __init__, você pode inicializar os atributos do objeto com os valores passados como parâmetros. Isso é comumente feito atribuindo os valores dos parâmetros aos atributos da classe usando a sintaxe self.atributo = valor.

## **Exemplos de Objetos**

````python
import math

class Circulo:

  def __init__(self, raio):
    self.raio = raio

  def calcular_area(self):
    return print( math.pi * self.raio ** 2 )

# Instancia
circulo1 = Circulo( 20 )

# Chamei o método
circulo1.calcular_area()
````

````python
class Livro:

  def __init__(self, titulo, autor, ano):

    # isinstance: verifica se o argumento ano é uma instância do tipo int
    if not isinstance(ano, int) or ano <= 0:
      raise ValueError('O ano deve ser um número inteiro positivo.')

    self.titulo = titulo
    self.autor = autor
    self.ano = ano

  def detalhes(self):
      return f'Título: {self.titulo} Autor: {self.autor} Ano: {self.ano}'
````

````python
livro1 = Livro('Silo', 'Hugh Howey', 2014)
livro1.detalhes()
````

````python
class ValidadorCPF:

    def __init__(self):
        self.lista_cpf_bloquear = ['123.456.789-10', '987.654.321-98', '111.222.333-44']

    def validar_cpf(self, cpf):

        # LOOP
        for cpf_na_lista in self.lista_cpf_bloquear:
            if cpf_na_lista == cpf:
                return True
        return False
````

````python
validador = ValidadorCPF()
````

````python
# Verificar o CPF
cpf_validar = '123.456.789-99'

if validador.validar_cpf(cpf_validar):
    print(f'O CPF {cpf_validar} está na lista.')

else:
    print(f'O CPF {cpf_validar} não está na lista.')
````

### **Buscando dados de uma API**

````python
import yfinance as yf
import pandas as pd
from datetime import datetime, timedelta
import time
````

````python
class YahooFinanceAPI:

  def __init__(self, simbolo):
    self.simbolo = simbolo
    self.dataframe = None

  def obter_dados_acao(self, meses=1):
    inicio = time.time()
    data_fim = datetime.today()
    data_inicio = data_fim - timedelta(days=meses*30)
    dados_acao = yf.download(self.simbolo, start=data_inicio, end=data_fim)
    fim = time.time()
    tempo_decorrido = fim - inicio
    print(f"Tempo decorrido para obter dados da API: {tempo_decorrido:.2f} segundos")
    self.dataframe = dados_acao
    return dados_acao

  def mostrar_estatisticas_basicas(self, meses=1):
    dados_acao = self.obter_dados_acao(meses)
    if not dados_acao.empty:
        print(f"Símbolo: {self.simbolo}")
        print(f"Número total de pontos de dados: {len(dados_acao)}")
        print(f"Máxima preço de fechamento: {dados_acao['Close'].max()}")
        print(f"Mínima preço de fechamento: {dados_acao['Close'].min()}")
        print(f"Média preço de fechamento: {dados_acao['Close'].mean()}")
    else:
        print("Não há dados disponíveis para o símbolo fornecido.")

  def mostrar_informacoes(self):
    if self.dataframe is None:
        print("DataFrame vazio. Leia um arquivo antes de usar esta função.")
    else:
        print(f"Número de linhas: {self.dataframe.shape[0]}")
        print(f"Número de colunas: {self.dataframe.shape[1]}")
        print(f"Colunas: {self.dataframe.columns}")
````

````python
yahoo_finance = YahooFinanceAPI("MGLU3.SA")
````

````python
yahoo_finance.mostrar_estatisticas_basicas(meses=3)
````

````python
yahoo_finance.mostrar_informacoes()
````

### **Criando um CRUD**

Famoso:
1. Create
2. Read
3. Updade
4. Delete

````python
import sqlite3
````

````python
class LivroDB:

  def __init__(self, db_name):
    self.conn = sqlite3.connect(db_name)
    self.c = self.conn.cursor()
    self.create_table()

  def create_table(self):
    Query = '''
      CREATE TABLE IF NOT EXISTS livros (
        id INTEGER PRIMARY KEY,
        titulo TEXT NOT NULL,
        autor TEXT NOT NULL,
        ano INTEGER NOT NULL)
    '''
    self.c.execute( Query )
    self.conn.commit()

  def adicionar_livro(self, titulo, autor, ano):
    self.c.execute("INSERT INTO livros (titulo, autor, ano) VALUES (?, ?, ?)", (titulo, autor, ano))
    self.conn.commit()

  def listar_livros(self):
    self.c.execute("SELECT * FROM livros")
    return self.c.fetchall()

  def buscar_livro_por_titulo(self, titulo):
    self.c.execute("SELECT * FROM livros WHERE titulo=?", (titulo,))
    return self.c.fetchall()

  def atualizar_livro(self, id, titulo, autor, ano):
    self.c.execute("UPDATE livros SET titulo=?, autor=?, ano=? WHERE id=?", (titulo, autor, ano, id))
    self.conn.commit()

  def deletar_livro(self, id):
    self.c.execute("DELETE FROM livros WHERE id=?", (id,))
    self.conn.commit()
````

````python
livro_db = LivroDB("livros.db")
````

````python
livro_db.adicionar_livro("Dom Casmurro", "Machado de Assis", 1899)
````

````python
livro_db.adicionar_livro("A Metamorfose", "Franz Kafka", 1915)
````

````python
print("Todos os livros:")
print(livro_db.listar_livros())
````

````python
print(livro_db.buscar_livro_por_titulo("Dom Casmurro"))
````

````python
livro_db.atualizar_livro(1, "Dom Casmurro", "Machado de Assis", 1900)
````

````python
livro_db.deletar_livro(2)
````
